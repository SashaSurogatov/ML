```python
import matplotlib.pyplot as plt 
import matplotlib.patches as patches 
import random 
import math 

q_size = math.sqrt(0.5) 

def generate_points(m, points): 
    for i in xrange(0, m): 
        x = random.uniform(0, 1) 
        y = random.uniform(0, 1) 
        points.append((x, y, x <= q_size and y <= q_size)) 

def generate_a(points): 
    top = max(points, key = lambda item: item[1] if item[2] else 0) 
    bottom = min(points, key = lambda item: item[1] if item[2] else 2) 
    left = min(points, key = lambda item: item[0] if item[2] else 2) 
    right = max(points, key = lambda item: item[0] if item[2] else 0) 

    if not top[2]: 
        return 0, 0, 0, 0 
    return left[0], top[1], bottom[1], right[0] 

m = 100
iteration = [] 
true_risk = [] 
for i in xrange(1, m): 
    points = [] 
    generate_points(i, points) 
    left, top, bottom, right = generate_a(points) 
    
    true_risk.append(q_size * q_size - (right - left) * (top - bottom)) 
    iteration.append(i) 

plt.plot(iteration, true_risk, 'r')
plt.xlabel('m')
plt.ylabel('true risk')
#plt.axis([0, m, 0, 0.1])
plt.show()
```
