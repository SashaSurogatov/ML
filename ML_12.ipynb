```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def read_data():
    x = []
    y = []
    with open('iris.data', 'r') as f:
        for line in f:
            if line == '\n':
                continue
            sepal_l, sepal_w, petal_l, petal_w, iris_class = line.split(',')
            x.append([sepal_l, sepal_w, petal_l, petal_w])
            if iris_class == 'Iris-setosa\n':
                y.append(1)
            elif iris_class == 'Iris-versicolor\n':
                y.append(2)
            else:
                y.append(3)
    return np.array(x).astype('float64'), np.array(y).astype('float64')

x, y = read_data()

tsne = TSNE(perplexity=30, learning_rate=10, n_iter=5000).fit_transform(x.copy())

pca = PCA(n_components=2).fit_transform(x.copy())

a, = plt.plot(tsne[:50, 0], tsne[:50, 1], '.r')
b, = plt.plot(tsne[50:100, 0], tsne[50:100, 1], '.g')
c, = plt.plot(tsne[100:150, 0], tsne[100:150, 1], '.b')
plt.legend([a, b, c], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.title('t-SNE')
plt.show()

plt.cla()
a, = plt.plot(pca[:50, 0], pca[:50, 1], '.r')
b, = plt.plot(pca[50:100, 0], pca[50:100, 1], '.g')
c, = plt.plot(pca[100:150, 0], pca[100:150, 1], '.b')
plt.legend([a, b, c], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.title('PCA')
plt.show()
```
